#  Data Preprossing

```python
# we can drop genres and timestamp as these columns are not required
df4=df3.drop(columns=['genres','timestamp'],axis='rows')
df4.head()
```

```python
df5=df4.pivot_table(index='userId',columns='title',values='rating')
df5.head()
```

```python
# as we havve only 610 users a lot of noise will be generated in the model which causes disturbance so we will drop few values
df6=df5.dropna(thresh=10,axis='columns')
df6.shape
```

```python
df7=df6.fillna(0)
df7
```

# Correlation

```python
correlation=df7.corr(method='pearson')
correlation
```

# Recommendation 

```python
def recommend(movie,rating):
    similarity_score=correlation[movie]*(rating-2.5)
    similarity_score=similarity_score.sort_values(ascending=False)[:15]
    return similarity_score
    
```

```python
recommend('2 Fast 2 Furious (Fast and the Furious 2, The) (2003)',5)
```

```python
action=[('2 Fast 2 Furious (Fast and the Furious 2, The) (2003)',5),('xXx (2002)',4),('(500) Days of Summer (2009)',2)]
recomendations=pd.DataFrame()
for movie,rating in action:
    recomendations=recomendations.append(recommend(movie,rating),ignore_index=True)
recomendations.sum().sort_values(ascending=False)[:15]
```
